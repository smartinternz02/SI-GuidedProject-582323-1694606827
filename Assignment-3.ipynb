# Outlier removal in Age using IQR

```python
q1=df.Age.quantile(0.25)
q3=df.Age.quantile(0.75)
print(q1)
print(q3)
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit=q3+1.5*IQR
upper_limit
```

```python
lower_limit=q3-1.5*IQR
lower_limit
```

```python
df=df[df.Age>lower_limit]
```

```python
df=df[df.Age<upper_limit]
```

```python
sns.boxplot(df.Age)
```

```python
df.shape
```

# replacong outliers in Fare using IQR

```python
sns.boxplot(df["Fare"])
```

```python
q1=df.Fare.quantile(0.25)
q3=df.Fare.quantile(0.75)
```

```python
print(q1)
print(q3)
```

```python
IQR=q3-q1
```

```python
whisker_width = 1.5
lower_whisker = q1 -(whisker_width*IQR)
upper_whisker = q3 + (whisker_width*IQR)
```

```python
df['Fare']=np.where(df['Fare']>upper_whisker,upper_whisker,
                    np.where(df['Fare']<lower_whisker,lower_whisker,df['Fare']))
```

```python
sns.boxplot(df['Fare'],data=df)
```
