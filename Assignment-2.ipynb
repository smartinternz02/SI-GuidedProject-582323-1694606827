```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
df
```

```python
df.head()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
sns.scatterplot(x='total',y='speeding',data=df)
```

```python
Inference: Total car crashes increases as the speeding increases.
```

```python
sns.distplot(df['total'])
plt.xlabel('Total crashes')
plt.ylabel('frequency')
```

```python
Inference : The maximum density of car crashes lies between 12 and 15
```

```python
corr=df.corr()
```

```python
sns.heatmap(corr,annot=True,cmap='YlGnBu')
```

```python
plt.figure(figsize=(17,6))
sns.boxplot(x='abbrev',y='total',data=df)
plt.xlabel('State Abbrevation')
plt.ylabel('Total crashes')
plt.xticks(rotation=90)
plt.show()
```

```python
Inference: The above box plot is helping us visualize the outliers of
    numeric variables i.e total crashes
```

```python
plt.figure(figsize=(17,8))
sns.barplot(x='abbrev',y='total',data=df)
plt.xlabel('State Abbrevation')
plt.ylabel('Total crashes')
plt.show()
```

```python
Inference: visualize categorical data, such as the state with the 
    highest total crashes.
```

```python
sns.regplot(x='ins_premium',y='ins_losses',data=df)
```

```python
plt.figure(figsize=(17,10))
sns.violinplot(x='abbrev',y='alcohol',data=df)
plt.xlabel('state abbrevation')
plt.ylabel('alcohol')
```

```python
Inference: The numeric variables such as alcohol,speeding can be 
    visualised across various states
```

```python
sns.scatterplot(x='alcohol',y='speeding',data=df)
```

```python
Inference: The alcohol level between 2 and 6 has highest recorded speeds
```

```python
sns.distplot(df['ins_premium'])
plt.xlabel('Insurance premium')
plt.ylabel('Density')
plt.show()
```

```python
Inference: above distribution plot visualizes the car insurance premiums
    according to its density
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='abbrev',y='ins_losses',color='green',data=df)
```

```python
Inference: from the above plot, we can say that comapny in state 'KY'
    has incurred most number of losses per insured driver
```

```python
sns.scatterplot(x='no_previous',y='total',marker='+',color='red',data=df)
plt.xlabel('% of drivers with no record of accidents')
plt.ylabel('Total car crashes')
```

```python
Inference:Its an constantly increasing plot i.e the  % of drivers
    who were not involved in any previous crashes is directly
    proportional to the total crashes.
```
